## Fetch stock data

### Subtask:
Download historical stock data for AMD.


**Reasoning**:
Import the yfinance library, define the ticker symbol for AMD, and download the historical stock data.



```python
import yfinance as yf

amd_ticker = 'AMD'
amd_data = yf.download(amd_ticker, period='1y')
```

**Reasoning**:
Display the first few rows of the downloaded DataFrame to verify the data.



```python
display(amd_data.head())
```

## Analyze stock data

### Subtask:
Calculate relevant financial indicators and perform basic analysis.


**Reasoning**:
Calculate the 20-day and 50-day simple moving averages and the daily percentage change of the 'Close' price and add them as new columns to the DataFrame.



```python
amd_data['SMA_20'] = amd_data['Close'].rolling(window=20).mean()
amd_data['SMA_50'] = amd_data['Close'].rolling(window=50).mean()
amd_data['Daily_Change'] = amd_data['Close'].pct_change()
display(amd_data.head())
display(amd_data.tail())
```

## Visualize stock data

### Subtask:
Plot the stock price and indicators.


**Reasoning**:
Plot the stock price and calculated indicators as instructed in the subtask.



```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

# Plot Close price, SMA_20 and SMA_50 on the first subplot
axes[0].plot(amd_data.index, amd_data['Close'], label='Close Price', color='blue')
axes[0].plot(amd_data.index, amd_data['SMA_20'], label='SMA 20', color='red')
axes[0].plot(amd_data.index, amd_data['SMA_50'], label='SMA 50', color='green')
axes[0].set_title('AMD Stock Price and Moving Averages')
axes[0].set_ylabel('Price')
axes[0].legend()
axes[0].grid(True)

# Plot Daily Change on the second subplot
axes[1].plot(amd_data.index, amd_data['Daily_Change'], label='Daily Change', color='purple')
axes[1].set_title('AMD Stock Daily Change')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Percentage Change')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
