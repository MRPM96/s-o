## Fetch stock data

### Subtask:
Download historical stock data for AMD.


**Reasoning**:
Import the yfinance library, define the ticker symbol for AMD, and download the historical stock data.



```python
import yfinance as yf

amd_ticker = 'AMD'
amd_data = yf.download(amd_ticker, period='1y')
```

**Reasoning**:
Display the first few rows of the downloaded DataFrame to verify the data.



```python
display(amd_data.head())
```

## Analyze stock data

### Subtask:
Calculate relevant financial indicators and perform basic analysis.


**Reasoning**:
Calculate the 20-day and 50-day simple moving averages and the daily percentage change of the 'Close' price and add them as new columns to the DataFrame.



```python
amd_data['SMA_20'] = amd_data['Close'].rolling(window=20).mean()
amd_data['SMA_50'] = amd_data['Close'].rolling(window=50).mean()
amd_data['Daily_Change'] = amd_data['Close'].pct_change()
display(amd_data.head())
display(amd_data.tail())
```

## Visualize stock data

### Subtask:
Plot the stock price and indicators.


**Reasoning**:
Plot the stock price and calculated indicators as instructed in the subtask.



```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

# Plot Close price, SMA_20 and SMA_50 on the first subplot
axes[0].plot(amd_data.index, amd_data['Close'], label='Close Price', color='blue')
axes[0].plot(amd_data.index, amd_data['SMA_20'], label='SMA 20', color='red')
axes[0].plot(amd_data.index, amd_data['SMA_50'], label='SMA 50', color='green')
axes[0].set_title('AMD Stock Price and Moving Averages')
axes[0].set_ylabel('Price')
axes[0].legend()
axes[0].grid(True)

# Plot Daily Change on the second subplot
axes[1].plot(amd_data.index, amd_data['Daily_Change'], label='Daily Change', color='purple')
axes[1].set_title('AMD Stock Daily Change')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Percentage Change')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

*   Key financial indicators, including the 20-day Simple Moving Average (SMA\_20), 50-day Simple Moving Average (SMA\_50), and Daily Percentage Change, were calculated and added to the dataset.
*   The closing price, SMA\_20, SMA\_50, and daily percentage change were visualized in separate subplots.

### Insights or Next Steps

*   The generated plots allow for visual identification of trends (using SMAs) and volatility (using Daily Change) in AMD's stock price over the past year.

# Task
Analyze the long NVDA vs short AMD thesis using historical stock data.

## Define the scope of the analysis

### Subtask:
Specify the time period and data sources for the analysis.


**Reasoning**:
Define the start and end dates for the analysis period and identify yfinance as the data source.



```python
import datetime

start_date = datetime.datetime.now() - datetime.timedelta(days=1*365)
end_date = datetime.datetime.now()

data_source = 'yfinance'

print(f"Analysis period: {start_date.date()} to {end_date.date()}")
print(f"Data source: {data_source}")
```

## Fetch stock data

### Subtask:
Download historical stock data for both NVDA and AMD.


**Reasoning**:
Download the historical stock data for NVDA and AMD using yfinance and display the first few rows of each DataFrame.



```python
import yfinance as yf

nvda_ticker = 'NVDA'
amd_ticker = 'AMD'

nvda_data = yf.download(nvda_ticker, start=start_date, end=end_date)
amd_data = yf.download(amd_ticker, start=start_date, end=end_date)

display(nvda_data.head())
display(amd_data.head())
```

## Analyze stock data

### Subtask:
Calculate relevant financial indicators and perform basic analysis for both stocks.


**Reasoning**:
Calculate the 20-day and 50-day simple moving averages and the daily percentage change for both NVDA and AMD stock data and display the head and tail of the dataframes to verify the results.



```python
nvda_data['SMA_20'] = nvda_data['Close'].rolling(window=20).mean()
nvda_data['SMA_50'] = nvda_data['Close'].rolling(window=50).mean()
nvda_data['Daily_Change'] = nvda_data['Close'].pct_change()

amd_data['SMA_20'] = amd_data['Close'].rolling(window=20).mean()
amd_data['SMA_50'] = amd_data['Close'].rolling(window=50).mean()
amd_data['Daily_Change'] = amd_data['Close'].pct_change()

display(nvda_data.head())
display(nvda_data.tail())
display(amd_data.head())
display(amd_data.tail())
```

## Compare performance

### Subtask:
Compare the performance of NVDA and AMD over the specified time period.


**Reasoning**:
Calculate the cumulative returns for both NVDA and AMD based on the 'Daily_Change' column and display the head and tail of the dataframes to verify.



```python
nvda_data['Cumulative_Return'] = (1 + nvda_data['Daily_Change']).cumprod() - 1
amd_data['Cumulative_Return'] = (1 + amd_data['Daily_Change']).cumprod() - 1

display(nvda_data.head())
display(nvda_data.tail())
display(amd_data.head())
display(amd_data.tail())
```

## Visualize results

# Task
Analyze the long NVDA vs short AMD thesis by creating a Profit and Loss backtest, assuming equal dollar amounts are invested in long NVDA and short AMD. Calculate and visualize the cumulative return and key performance metrics such as total return, annualized return, maximum drawdown, and Sharpe ratio. Summarize the backtest results.

## Calculate daily returns for nvda and amd

### Subtask:
Use the 'Daily_Change' column to calculate the daily returns for both stocks.


## Calculate the daily return of the strategy

### Subtask:
Assuming equal dollar amounts are invested in long NVDA and short AMD, calculate the daily return of the combined strategy.


**Reasoning**:
Calculate the daily return of the combined strategy by taking the 'Daily_Change' of NVDA and subtracting the 'Daily_Change' of AMD, and store the result in the 'Strategy_Daily_Return' column of the nvda_data DataFrame. Then, display the head and tail of the nvda_data DataFrame to verify the new column and calculations.



```python
nvda_data['Strategy_Daily_Return'] = nvda_data['Daily_Change'] - amd_data['Daily_Change']
display(nvda_data.head())
display(nvda_data.tail())
```

## Calculate the cumulative return of the strategy

### Subtask:
Calculate the cumulative return of the strategy over the entire period.


**Reasoning**:
Calculate the cumulative return of the strategy by taking the cumulative product of (1 + daily return) and subtracting 1, then display the head and tail of the dataframe to verify the calculation.



```python
nvda_data['Strategy_Cumulative_Return'] = (1 + nvda_data['Strategy_Daily_Return']).cumprod() - 1
display(nvda_data.head())
display(nvda_data.tail())
```

## Visualize the cumulative return of the strategy

### Subtask:
Visualize the cumulative return of the strategy over the entire period.


**Reasoning**:
Plot the cumulative return of the strategy as instructed in the subtask.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(nvda_data.index, nvda_data['Strategy_Cumulative_Return'], label='Strategy Cumulative Return', color='blue')
plt.title('Cumulative Return of Long NVDA vs Short AMD Strategy')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()
```

## Calculate key performance metrics

### Subtask:
Calculate key performance metrics such as total return, annualized return, maximum drawdown, and Sharpe ratio for the strategy.


**Reasoning**:
Calculate and print the total return, annualized return, maximum drawdown, and Sharpe ratio for the strategy using the provided formulas and assumptions.



```python
total_return = nvda_data['Strategy_Cumulative_Return'].iloc[-1]

years = (nvda_data.index[-1] - nvda_data.index[0]).days / 365.25
annualized_return = (1 + total_return)**(1 / years) - 1

nvda_data['Running_Max'] = nvda_data['Strategy_Cumulative_Return'].cummax()
nvda_data['Daily_Drawdown'] = nvda_data['Running_Max'] - nvda_data['Strategy_Cumulative_Return']
max_drawdown = nvda_data['Daily_Drawdown'].max()

risk_free_rate = 0
trading_days_in_year = 252
sharpe_ratio = (nvda_data['Strategy_Daily_Return'].mean() - risk_free_rate) / nvda_data['Strategy_Daily_Return'].std() * (trading_days_in_year**0.5)

print(f"Total Return: {total_return:.4f}")
print(f"Annualized Return: {annualized_return:.4f}")
print(f"Maximum Drawdown: {max_drawdown:.4f}")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
```
