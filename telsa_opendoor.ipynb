## Fetch tesla stock data

### Subtask:
Download historical data for Tesla (TSLA) for the last year using `yfinance`.


**Reasoning**:
Download historical data for Tesla (TSLA) for the last year using `yfinance`.



```python
import yfinance as yf
from datetime import date, timedelta

today = date.today()
one_year_ago = today - timedelta(days=365)
ticker = 'TSLA'
tsla_data = yf.download(ticker, start=one_year_ago, end=today)
```

**Reasoning**:
Display the first few rows of the downloaded data to verify the download.



```python
display(tsla_data.head())
```

## Calculate daily returns

### Subtask:
Compute the daily percentage change in stock price.


**Reasoning**:
Calculate the daily percentage change of the 'Close' price and store it in a new column named 'Daily Returns'.



```python
tsla_data['Daily Returns'] = tsla_data['Close'].pct_change()
display(tsla_data.head())
```

## Calculate moving averages

### Subtask:
Calculate a short-term and a long-term moving average for the closing price.


**Reasoning**:
Calculate the 50-day and 200-day moving averages for the 'Close' price and display the updated dataframe.



```python
tsla_data['50-Day Moving Average'] = tsla_data['Close'].rolling(window=50).mean()
tsla_data['200-Day Moving Average'] = tsla_data['Close'].rolling(window=200).mean()
display(tsla_data.head())
```

## Visualize the data

### Subtask:
Create a plot showing the daily returns and the moving averages.


**Reasoning**:
Generate the plot showing the daily returns and moving averages as requested in the subtask instructions.



```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

axes[0].plot(tsla_data.index, tsla_data['Daily Returns'], label='Daily Returns', color='blue')
axes[0].set_title('Tesla Daily Returns')
axes[0].set_ylabel('Daily Return')
axes[0].grid(True)

axes[1].plot(tsla_data.index, tsla_data['Close'], label='Close Price', color='black')
axes[1].plot(tsla_data.index, tsla_data['50-Day Moving Average'], label='50-Day Moving Average', color='red')
axes[1].plot(tsla_data.index, tsla_data['200-Day Moving Average'], label='200-Day Moving Average', color='green')
axes[1].set_title('Tesla Stock Price and Moving Averages')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Price')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
